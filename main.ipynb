```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
from scipy.signal import find_peaks

data = np.array( pd.read_csv('正坐') )
data = data[ 0 : ,0] # 選取要的範圍

peak_list = find_peaks( data , height = 80 , distance = 250 )[0]  # 每一下心跳的波鋒之時間
peak = np.zeros((len( peak_list )-1,340), dtype=int)
data = data.tolist() # float64 -> float
peak_list = peak_list.tolist()

for i in range( 0 , len( peak_list )-1 ) :
    k=0
    for j in range ( peak_list[i]-170 , peak_list[i]+170 ) :
        num = int(data[j])
        peak[i][k] = num
        k+=1
        
peak = np.mean( peak , axis=0 )
```
